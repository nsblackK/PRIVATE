<h1> Binomial probability distribution <h1>

```r
#probability of head in a single coin flipping
rbinom(10,1,0.5)
```

```r
#probability of head in 10 coin flipping
rbinom(10,10,0.5)
```

```r
#probability of 3 heads in 10 flips of a coin
pbinom(3,10,0.5)
```

```r
pbinom(5,10,0.5)
```

```r
#reverse of pbinom
qbinom(0.17,10,0.5)
```

```r
#deaw the prob dist of the case of flipping 10 coins 
#and looking at the probabilities of various number ofhead
n<-c(1:10)
D=dbinom(n,10,0.5)
print(D)
barplot(D)
```

```r
#lets visualize prob of 5 or less heads in 10 flips of coins
library(visualize)
visualize.binom(5,10,0.5)
```

```r
#probability of 5 or more heads in fliping 10 coins
visualize.binom(5,10,0.5,section="upper")
```

```r
#prob oh head of heads between 4 nd 5
visualize.binom(c(4,5),10,0.5,section="bounded")
```

```r
visualize.binom(c(1,9),10,0.5,section="tails")
```
