```r
#draw t distribution for diff degree of freedom
q <- seq(-4,4,by=0.1)
q
```

```r
dt(q,19)
plot(q,dt(q,19),type="l",lty="solid",xlab="t",ylab="f(t)")
lines(q,dt(q,9),type="l",lty="dashed")
lines(q,dt(q,4),type="l",lty="dotted")
```
