```python
import numpy 

canvas=numpy.ndarray(shape=(8,8), dtype=int)
#print(canvas)
```

```python
#reset
def reset():
    for i in range(len(canvas)):
        canvas[i]=0
    #print(canvas)
reset()
#print(canvas)
```

```python
def check(a,b):
    l=0
    for i in range(8):
        #print("a")
        if canvas[a,i]!=0 or canvas[i,b]!=0:
            l+=1
            
    c=canvas.diagonal(b-a)
    for j in c:
        if j!=0:
            print("there")
            l+=1
    d=numpy.fliplr(canvas).diagonal(7-b)
    print(d)
    for j in d:
        if j!=0:
            print("here",j)
            l+=1
                    
    if l>0:
        print(l)
        return False
    else:
        return True
        
check(7,0)
```
