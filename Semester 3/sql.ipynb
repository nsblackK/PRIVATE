```python
EMP=[]
import random
#random fill
def fill(A):
    for i in range(5):
        A.append(random.randint(0, 100))
        A.sort()

#insert one element    
def insert(A,t):
    A.append(t)
    A.sort()

#delete one element
def delete(A,t):
    for i in range(len(A)):
        if A[i] == t:
            del A[i]

#delete all
def deleteall(A):
    A.clear()

#display all
def disp(A):
    print(A)
    
#count
def count(A):
    print(len(A))
    
#display 1
def dispone(A,t):
    for i in range(len(A)):
        if A[i] >= t:
            print(A[i])
            
#max
def maxi(A):
    m=A[0]
    for i in range(len(A)):
        if m < A[i]:
            m=A[i]
    print(m)

#min    
def mini(A):
    m=A[0]
    for i in range(len(A)):
        if m > A[i]:
            m=A[i]
    print(m)
    
#duplicate    
def dupli(A):
    
    for i in range(len(A)):
        for j in range(i+1,len(A)):
            if A[i]==A[j]:
                print(A[i])
#remove duplicate    
def Rdupli(A):
    l= len(A)
    for i in range(l-1):
        if A[i]==A[i+1]:
            del A[i+1]
            break
        
```
