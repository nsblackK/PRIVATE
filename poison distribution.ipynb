```r
#on a booking counter 3.6 people care coming every 10 mins on an avg.
#what is the probability of getting 7 ppl in 10 mins
dpois(7,3.6)
```

```r
ppois(7,3.6)-ppois(6,3.6)
```

```r
#plot prob dist for the above example
a<-c(0:10)
dpois(a,3.6)
barplot(dpois(a,3.6))
library (visualiza)
visualize.pois(7,3.6)
```

```r
#probability of 5 or more
library(visualize)
visualize.pois(5,3.6,section="upper")
```

```r
library(visualize)
visualize.pois(c(0,8),3.6, section = "tails")
```

```r
visualize.pois(c(7,7),3.6,section="bounded")
```
