```python
"""1. A client has enrolled for python programming online course, course fees is 5000 but the
company offers discount of 15% for clients below the age group 22, 10% discount
between age group 23 to 25, 5% discount for age group above 25. Calculate the amount
to be paid by the client while enrolling for the course. Meanwhile the Website wishes to
display the existing courses provided by the company to the clients to gain more profit.
Write an effective program in python to make sure companies requirements are met."""
```

```python
import threading
import time

def calculate():
    p=5000
    age=int(input("enter age "))
    e.set()
    if age<22:
        pay = p-p*0.15
    elif age>23 and age<25:
        pay = p-p*0.10
    else:
        pay=p-p*0.05
    time.sleep(0.1)
    e.clear()
    print("You, my son will have to pay",pay)

def promotion():
    e.wait()
    while e.isSet():
        print("learn R with us")
        print("learns 3d max in 2 months!")
        print("learn this")
        print("learn that")

e = threading.Event()    
t1=threading.Thread(target=calculate)
t2=threading.Thread(target=promotion)
t1.start()
t2.start()
```

```python
"""2. An organization wants to start online donation program were it would collect items
from doners and store it in its repository and distribute to two needy groups which have
10 members each. But the organization follows a rule only when item count reaches 20
it will start distributing the items to the 2 needy groups where first 10 items will be
picked by group 1 and rest will be picked by group 2 and assign to the members in the
group. Finally, both the groups should display the member names and the assigned
items in the organization website. (Items can be taken as user input). Write an effective
python program to accomplish the task of the organization."""
```

```python
import threading
import time
import queue

g1=["sam","tom","hardy","matt","joe","jos","jack","dani","burt","mac"]
g2=["samurai","tomy","haunt","matty","ethen","josua","jacky","danial","burto","jeff"]

q=queue.Queue()

def donate():
    while q.qsize()<20:
        i=input("enter item name ")
        q.put(i) 
    e.set()
    
def distribute():
    e.wait()
    while e.isSet():
        while not q.empty():
            for i in g1:
                print(i,"gets",q.get())
            for j in g2:
                print(j, "gets", q.get())
                
e = threading.Event()     
t1=threading.Thread(target=donate)
t2=threading.Thread(target=distribute)
t1.start()
t2.start()
t1.join()
t2.join()
print("done")

            
            
```

```python
"""3. An Organization wanted to start online donation program for puppies, it already has a
pool of details which consists of pets age and owner name is the form of dictionary. The
organization wants to donate to 2 Adoption groups randomly such that the groups can
save the details in their repository and hand over the puppies when people want to
adopt one. The organization wants to run the program effectively online with very less
time and resource utilization."""
```

```python
import threading
import time
import queue

q=queue.Queue()

details={2:"Arya",3:"Ram",1:"Sham",1.5:"Rahul",2.3:"Arya",3.2:"Ram",
         1.1:"Sham",3.5:"Rahul",2.22:"Arya",3.35:"Ram",15:"Sham",
         1.55:"Rahul",2.35:"Arya",1.25:"Ram",0.1:"Sham",4.5:"Rahul"}

def inp():
    for i in details.items():
        q.put({i})

def grp1():
    while not q.empty():
        print("group 1: ",q.get())
        
def grp2():
    while not q.empty():
        print("group 2: ",q.get())
        

t1=threading.Thread(target=inp)
t2=threading.Thread(target=grp1)
t3=threading.Thread(target=grp2)
t1.start()
t2.start()
t3.start()
```

```python
"""4. An Organization wanted to start online donation program for puppies, it already has a
pool of details which consists of pets age and owner name is the form of dictionary. The
organization wants to donate to 2 Adoption groups such that the groups can save the
details in their repository and hand over the puppies when people want to adopt one.
(Since the details of puppies are not sorted based on age the organization wants to give
more priority to the smaller age group puppies and send their details first). The
organization wants to run the program effectively online with very less time and
resource utilization."""
```

```python
import threading
import time
import queue

q=queue.PriorityQueue()

details={2:"Arya",3:"Ram",1:"Sham",1.5:"Rahul",2.3:"Arya",3.2:"Ram",
         1.1:"Sham",3.5:"Rahul",2.22:"Arya",3.35:"Ram",15:"Sham",
         1.55:"Rahul",2.35:"Arya",1.25:"Ram",0.1:"Sham",4.5:"Rahul"}

def inp():
    for i in details.items():
        q.put(i)

def grp1():
    while not q.empty():
        print("group 1: ",q.get())
        
def grp2():
    while not q.empty():
        print("group 2: ",q.get())
        

t1=threading.Thread(target=inp)
t2=threading.Thread(target=grp1)
t3=threading.Thread(target=grp2)
t1.start()
t2.start()
t3.start()
```
