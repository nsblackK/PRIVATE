```python
import socket
s=socket.socket()
print("Socket connected")

host="data.pr4e.org"
port=80

s.connect((host,port))
print("connected")


msg="GET http://data.pr4e.org/intro.txt HTTP/1.0\r\n\r\n".encode()
s.send(msg)

data=s.recv(1024)
print(data.decode())

```

```python
import socket
s=socket.socket()
print("Socket connected")

host="onepiecegold.com"
port=80

s.connect((host,port))
print("connected")


msg="GET http://www.onepiecegold.com/category/theories/NEWS.txt HTTP/1.0\r\n\r\n".encode()
s.send(msg)

while True:
    data=s.recv(1024)
    if len(data)<1:
        break
    print(data.decode())
```
