# SET 2: United States - Crime Rates - 1960 - 2014

```python
crime_rates = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
crime_rates.head()
```

```python
#Q1
crime_rates.info()
```

```python
#Q2
crime_rates.Year = pd.to_datetime(crime_rates.Year,format='%Y')
crime_rates.Year
```

```python
#Q
crime_rates = crime_rates.set_index('Year',drop=True)
crime_rates
```

```python
crimes = crime_rates.resample('10AS').sum()

population = crimes['Population'].resample('10AS').max()
print(population)
crimes['Population'] = population
crimes
```

```python
crimes.idxmax()
```
