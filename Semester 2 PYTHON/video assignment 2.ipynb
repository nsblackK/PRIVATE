```python
import threading 
import time

def dev():
    print("thinking idea")
    time.sleep(2)
    q.acquire()
    print("making graphics")
    print("making game logic")
    q.notifyAll()
    q.release()
    
def salse():
    q.acquire()
    q.wait()
    print("selling 1st copy")
    q.release()
    
def salse2():
    q.acquire()
    q.wait()
    print("selling 2nd copy")
    q.release()
    
q=threading.Condition()
t1=threading.Thread(target=dev)
t2=threading.Thread(target = salse)
t3=threading.Thread(target = salse2)
t1.start()
t2.start()
t3.start()
```

```python
import socket
s=socket.socket()

host="www.onepiecegold.com"
port=80

s.connect((host,port))

req="GET http://www.onepiecegold.com/NEWS.txt HTTP/1.0\r\n\r\n".encode()
s.send(req)

while True:
    data=s.recv(1024)
    if len(data)<1:
        break
    print(data.decode())
```
