```python
#a function with yield is a generator
def display():
    yield 1
    yield 2
c=display()
print(c)
#print(next(c)) #accessing the data

for i in c:
    print(i)
```

```python
#normal inefficient way
def cube():
    cubeList=[]
    for i in range(1,501):
        cubeList.append(i**3)
    return cubeList

d=cube()
print(d)
```

```python
#special efficien way using yield(generator)
def cube():
    cubeList=[]
    for i in range(1,501):
        yield(i**3)
    

d=cube()
for i in d:
    print(i)
```

```python
sal=[35000,40000,50000,55000,70000,50000,60000,46000,75000,36000]

def tax_cal(salary):
    for i in salary:
        if(i*12>500000):
            tax=5/100*(i*12)
            yield (tax)
        else:
            tax=0
            yield (tax)
c=tax_cal(sal)
for i in c:
    if(i==0):
        print("No Tax")
    else:
        print(i)
```
