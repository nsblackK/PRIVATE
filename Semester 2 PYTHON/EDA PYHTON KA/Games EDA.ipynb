```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
games=pd.read_csv('vgsales.csv')
games.head()
```

```python
games.describe()
```

```python
print(games.Platform.unique())
print(games.Genre.unique())
print(games.Publisher.unique())
```

```python
sns.set_style("darkgrid")
import matplotlib
```

```python
#what is the Game Sales Across the World Throughout the Years?
sns.set(font_scale=2)
plt.figure(figsize=(20,12))

plt.plot(games.groupby('Year').Global_Sales.agg(['sum']))

plt.xlabel('Years')
plt.ylabel('Global Sale')
plt.title("Game Sales Across the World Throughout the Years");
```

```python
plt.figure(figsize=(35,15))
plt.title("On an average, which Genre sold more copies globally?")

sns.barplot(x='Genre', 
                y='Global_Sales', 
               data=games);
```

```python

#most games released, are in which Genre?
plt.figure(figsize=(40,15))
plt.hist(games.Genre);
```

```python
plt.figure(figsize=(20,12))
plt.title("Sales comparision in North America and Europe based on genre of game")

sns.scatterplot(x='EU_Sales', 
                y='NA_Sales', 
                hue='Platform', 
                s=100,
               data=games);
```

```python
plt.figure(figsize=(20,12))
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title("Game Sales comparision in different regions Throughout the Years");



plt.plot(games.groupby('Year').NA_Sales.agg(['sum']))
plt.plot(games.groupby('Year').EU_Sales.agg(['sum']))
plt.plot(games.groupby('Year').JP_Sales.agg(['sum']))

plt.legend(['North America','Europe', "Japan"]);

```
