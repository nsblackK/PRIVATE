```python
"""Topics covered in this code are:
urllib
Sending Email
Web Services :JSON,API

PROBLEM: Your Boss wants the link to all the universities of the country of his choise. So Extract 
the university details from the given api and store it in a json file
and send an email with the json file as an attachment"""



import urllib.request,urllib.parse
import json
UniDic={}

furl="http://universities.hipolabs.com/search?"
country=input("Enter country name: ")
apurl=urllib.parse.urlencode({"country":country})
url=furl+apurl

data=urllib.request.urlopen(url)
jdata=data.read().decode()
d=json.loads(jdata)

k=0
for i in d:
    UniDic[(d[k]['name'])]=(d[k]['web_pages'][0])
    k+=1


d=json.dumps(UniDic, indent=3, sort_keys=True)


f=open('UniversityState.json','w')
f.write(d)
f.close()
```

```python
"""Topic: Queues nd threading conditions

PROBLEM: take book recomendations from different users, nd suggest one book when asked"""

import threading
import time
import queue

         
e=threading.Condition()
q=queue.Queue()

def prod():
    c=0
    e.acquire()
    while c<10:
        b=input("enter book name")
        c+=1
        q.put(b)
    e.notifyAll()
    e.release()
    
def cons():
    e.acquire()
    e.wait()
    
    print(q.get())
    e.release()


t1=threading.Thread(target=prod)
t2=threading.Thread(target=cons)
t1.start()
t2.start()
```

```python
"""TOPIC: sockets

PROBLEM: Extract data from Christ university website"""

import socket
s=socket.socket()
print("Socket connected")

host="christuniversity.in"
port=80

s.connect((host,port))
print("connected")


msg="GET https://lavasa.christuniversity.in/courses/Main%20Campus/data-science,-economics-and-analytics/computer-science/bsc-(data-science) HTTP/1.0\r\n\r\n".encode()
s.send(msg)

while True:
    data=s.recv(1024)
    if len(data)<1:
        break
    print(data.decode())
```
