```r
df<-read.csv("Q1.csv")
head(df)
```

```r
summary(df)
```

```r
library(ggplot2)
ggplot(df,aes(month))+geom_bar(aes(fill= month))
```

```r
plot(x = df$FFMC, y = df$DMC, xlab = "FFMC",ylab = "DMC", main = "FFMC vs DMC")
```

```r
library(ggplot2)

ggplot(df,aes(x=day, y=rain)) + geom_jitter(color="blue", size=2, alpha=0.3)
```

```r
library(dplyr)
df %>%
 ggplot(aes(month, wind)) +
 geom_boxplot(fill = "yellow", colour = "black") +

 theme_minimal() 
```

```r
library(ggpubr)
ggscatter(df, x = "wind", y = "RH", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "pearson",
          xlab = "wind", ylab = "RH")
```

```r
#C2
df %>%
  group_by(month) %>%
  summarise(X,Y,temp>mean(temp))
```

```r
#C1
df %>%
  group_by(month) %>%
  summarise(FFMC=mean(FFMC))
```
