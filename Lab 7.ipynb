```r
library(readxl)
data <- read_excel("C:/Users/DELL/Downloads/titanic_train.xlsx")
print(head(data))
```

```r
#numerical columns are: age, fare
#categorical columns are: pclass, sex, embarked, survived
```

```r
#NUMERICAL VS NUMERICAL
plot(x = data$age, y = data$fare,xlab = "age",ylab = "fare", main = "fare vs age")
```

```r
#NUMERICAL VS CATEGORICAL
library(ggplot2)
ggplot(data,aes(x=fare,fill=pclass)) + geom_density(alpha=0.5) + labs(title="Fare Distribution")

ggplot(data,aes(x=pclass, y=fare)) + geom_jitter(color="blue", size=2, alpha=0.3)
```

```r
#CATEGORICAL vs CATEGORICAL

ggplot(data,aes(x=pclass, fill = sex)) + geom_bar(position = position_dodge(preserve = "single"))
```

```r
ggplot(data,aes(x=pclass,fill=sex))+geom_bar(position="stack")
```
